```python
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
dirs = {">":0,"v":1,"<":2,"^":3,"#":-1,".":-2}
init_snow_storms = {}
board = []
for line in lines:
    tmp = []
    for char in line:
        tmp.append(dirs[char])
    board.append(tmp)
board = np.array(board)
for i in range(4):
    init_snow_storms[i] = np.zeros(board.shape)
    init_snow_storms[i][board==i] = 1
walls = np.zeros(board.shape)
walls[board==-1] = 1
```

```python
print(init_snow_storms[0])
tmp = np.roll(init_snow_storms[0],1,axis=1)
tmp[:,1] = tmp[:,-1]
tmp[:,-1] = 0
# print(tmp)
# print("snow 0")
# print(snow_storms[0])
# print(roll_one(snow_storms[0],1,1,1,-1))
# print("snow 1")
# print(snow_storms[1])
# print(roll_one(snow_storms[1],0,1,1,-1))
# print("snow 2")
# print(snow_storms[2])
# print(roll_one(snow_storms[2],1,-1,-2,0))
# print("snow 3")
# print(snow_storms[3])
# print(roll_one(snow_storms[3],0,-1,-2,0))

dirs = {">":0,"v":1,"<":2,"^":3,"#":-1,".":-2}

def roll_one(array,axis,amount,start,end):
    tmp = np.roll(array,amount,axis=axis)
    if axis == 1:
        tmp[:,start] = tmp[:,end]
        tmp[:,end] = 0
    elif axis == 0:
        tmp[start,:] = tmp[end,:]
        tmp[end,:] = 0
    return tmp

def update_snow_storms(snow_storms):
    new_snow_storms = {}
    new_snow_storms[0] = roll_one(snow_storms[0],1,1,1,-1)
    new_snow_storms[1] = roll_one(snow_storms[1],0,1,1,-1)
    new_snow_storms[2] = roll_one(snow_storms[2],1,-1,-2,0)
    new_snow_storms[3] = roll_one(snow_storms[3],0,-1,-2,0)
    return new_snow_storms
```

```python
found_end = False
initial_board = init_snow_storms[0] + init_snow_storms[1] + init_snow_storms[2] + init_snow_storms[3] + walls
snow_storms = init_snow_storms
print(initial_board)
start = np.where(initial_board[:,:] == 0)
print(start)
start_t = (start[0][0],start[1][0])
points = set([(start[0][0],start[1][0])])
print("starting at ",points)
end = np.where(initial_board[:,:] == 0)
end_point = (end[0][-1],end[1][-1])
print("ending at",end_point)
round_count = 0
while not found_end:
    snow_storms = update_snow_storms(snow_storms)
    
    current_board = snow_storms[0] + snow_storms[1] + snow_storms[2] + snow_storms[3] + walls
    #print(current_board)
    new_points = set()
    for point in points:
        for d in [(0,0),(-1,0),(1,0),(0,-1),(0,1)]:
            new_point = (d[0]+point[0],d[1]+point[1])
            if new_point[0] == end_point[0] and new_point[1] == end_point[1]:
                found_end = True
            if current_board[new_point] == 0:
                new_points.add(new_point)
    
    points = new_points
    print("round is",round_count)
    print("next round has",len(points),"point(s)")
    round_count += 1
print(round_count)
```

```python
# back track for lunch
start_t = (start[0][0],start[1][0])
end_t = (end[0][-1],end[1][-1])
points = set([end_t])
print("starting at ",points)
end_point = (start_t)
found_end = False
# round_count = 0

while not found_end:
    snow_storms = update_snow_storms(snow_storms)
    
    current_board = snow_storms[0] + snow_storms[1] + snow_storms[2] + snow_storms[3] + walls
    #print(current_board)
    new_points = set()
    for point in points:
        for d in [(0,0),(-1,0),(1,0),(0,-1),(0,1)]:
            new_point = (d[0]+point[0],d[1]+point[1])
            if new_point[0] == end_point[0] and new_point[1] == end_point[1]:
                found_end = True
            if new_point[0] < current_board.shape[0] and new_point[1] < current_board.shape[1] and current_board[new_point] == 0:
                new_points.add(new_point)
    
    points = new_points
    print("round is",round_count)
    print("next round has",len(points),"point(s)")
    round_count += 1
```
