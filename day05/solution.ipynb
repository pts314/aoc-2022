```python
board_data,moves = data.split("\n\n")
```

```python
labels = board_data.split("\n")[-1]
```

```python
board = {}
loc = {}
lines = board_data.split("\n")
for i in range(1,len(labels),4):
    board[int(labels[i])] = []
    loc[i] = int(labels[i])
for i in range(len(lines)-2,-1,-1):
    #print(i)
    for j in range(1,len(labels),4):
        char = lines[i][j]
        if char != " ":
            board[loc[j]].append(char)
            #print("added",char)
    #|print(lines[i])
```
